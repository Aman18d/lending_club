    ## Based on above we can drop more columns as the data looks not relevant 

```python
cloumn_remove_1 = ["desc","mths_since_last_delinq","mths_since_last_record","next_pymnt_d"]
df_new.drop(labels = cloumn_remove_1, axis =1, inplace=True)
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 50 ##Check this not working
display(df_new)
display(df_new.describe())
```

    ##Employee length and term columns can be used in betterway if they have only numericals also 1000+ records are null in emp length column also other columns also like % in interest rate to be remove.

```python
df_new.emp_length.fillna('0', inplace= True)
df_new['emp_length']=df_new.emp_length.str.extract('(\d+)')
df_new['int_rate'] = df_new['int_rate'].str.rstrip('%')

df_new['revol_util'] = df_new['revol_util'].str.rstrip('%')
df_new.describe() ##The above columns after removing char are not showing in stats so it a char field need to change to numeric
```

```python
df_new.info()
```

```python
col_list = ["emp_length","int_rate","revol_util"]
df_new[col_list] = df_new[col_list].apply(pd.to_numeric)
df_new.head(5)
```

# Univariate Anlysis lets start analyzing the above data and see the individual set of column values and there content 

```python
print("###### Check the different terms #######")

print(df_new.term.value_counts())

print("###### Check the different Verification status #######")    
print(df_new.verification_status.value_counts())
print("###### Check the different Home OWNER #######")    
print(df_new.home_ownership.value_counts())
print("###### Check the different LOAN Status #######")    
print(df_new.loan_status.value_counts())
print("###### CHECK THE DIFERENT PURPOSE OF LOAN ######")
print(df_new.purpose.value_counts())
```

## Lets try to draw some conclusion b/w purpose of loan and the status what type of loan paid first

```python
plt.figure(figsize=(14,8),facecolor='y')
sns.set_style("darkgrid")  # white, dark, whitegrid, darkgrid, ticks
#Plot X and Y axis
ax = sns.countplot(y="purpose",data=df_new,hue='loan_status',palette='OrRd_r') ##     RdYlBu
ax.set_title('COnclusion B/W Loan status and Purpose',fontsize=14,color='g')
ax.set_ylabel('Purpose',fontsize=14,color = 'b')
ax.set_xlabel('Application Status',fontsize=14,color = 'y')      
plt.show()
```

    ## With above we can conclude that max people taken loan for DEBT_consoLIDATION AND CC Bill Payment and their Fully Paid % is also high but in Debt_Consolidation the charge off is also high that means people default there also small Business aslo percentage wise is higher

###Also from the above 128 Grid we try to find some conclusion from Loan status and Type of ownership as rent and mortgage is huge %

```python
plt.figure(figsize=(14,6),facecolor='y')
ax = sns.countplot(y="home_ownership",data=df_new,hue='loan_status',palette='OrRd_r')
ax.set_title('Home Ownership VS Loan Status',fontsize=14,color='g')
ax.set_xlabel('Loan Status',fontsize=14,color = 'r')
ax.set_ylabel('Home Ownership',fontsize=19,color = 'g')      
plt.show()
```

##Above shows that pergentage wise those who mortgage make more Charge off
## Also we have seen there are 2 terms above 36 and 60

```python
plt.figure(figsize=(20,10),facecolor='y')
ax = sns.countplot(y="term",data=df_new,hue='loan_status', linewidth=2, edgecolor=sns.color_palette("dark", 3)) ##palette='OrRd_r'
ax.set_title('Term VS STATUS',fontsize=14,color='g')
ax.set_xlabel(' Term',fontsize=14,color = 'r')
ax.set_ylabel(' Application Status',fontsize=14,color = 'g')      
ax.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(25,10),facecolor='y')
ax = sns.countplot(x="annual_inc",data=df_new,hue='loan_status',linewidth=2, edgecolor=sns.color_palette("dark", 3))
ax.set_title('INCOME VS STATUS',fontsize=14,color='g')
ax.set_xlabel(' Income',fontsize=14,color = 'r')
ax.set_ylabel(' Application Status',fontsize=14,color = 'g')      
ax.legend(bbox_to_anchor=(1, 1))
plt.show()
```

## Its Clear evident that long term loans default More and High salaried tend to default minimum and its maximum at mid tear and max between 25% to 50% salary band.
# So from this Analysis we can define that Long term Loans and those who Mortgage and Those who either take loan for Debt Consolidation or CC tend to make more defaults.
